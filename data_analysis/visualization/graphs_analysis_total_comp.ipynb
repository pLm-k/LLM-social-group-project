```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

file_path = 'path/to/data_analysis' #------------------change path to data_analysis dir here-------------!
data = pd.read_csv(file_path+'/results/results.csv', sep='\t')
data_preproc = pd.read_csv(file_path+'/results/results_preproc.csv', sep='\t')

years = data['year']

bar_width = 0.3
bar_pos1 = np.arange(len(years))
bar_pos2 = bar_pos1 + bar_width
```

```python
# Plotting a bar graph
plt.bar(bar_pos1, data['num_comments'], width=bar_width, label='data')
plt.bar(bar_pos2, data_preproc['num_comments'], width=bar_width, label='data preprocessed')
plt.xticks(bar_pos1 + bar_width / 2, years)
plt.legend()
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
# Adding labels and title
plt.xlabel('years')
plt.ylabel('num_comments')
plt.title('num_comments per year')

# Display the plot
plt.show()
```

```python
# Plotting a bar graph
plt.bar(bar_pos1, data['num_sentences'], width=bar_width, label='data')
plt.bar(bar_pos2, data_preproc['num_sentences'], width=bar_width, label='data preprocessed')
plt.xticks(bar_pos1 + bar_width / 2, years)
plt.legend()
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
# Adding labels and title
plt.xlabel('years')
plt.ylabel('num_sentences')
plt.title('num_sentences per year')

# Display the plot
plt.show()
```
