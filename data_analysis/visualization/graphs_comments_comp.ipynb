```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

file_path = 'path/to/data_analysis' #------------------change path to data_analysis dir here-------------!

years = ['2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019'] #!--------------------------------set desired years--------------------------!

df_years = []
df_years_preproc = []

#csvs tab seperated!
for year in years:
    df_years.append(pd.read_csv(file_path+f'/results/result_comments_{year}.csv', sep='\t'))
    df_years_preproc.append(pd.read_csv(file_path+f'/results/result_comments_{year}_preproc.csv', sep='\t'))

df_total = pd.read_csv(file_path+f'/results/result_comments_total.csv', sep='\t')
df_total_preproc = pd.read_csv(file_path+f'/results/result_comments_total_preproc.csv', sep='\t')

n_subs = 20

bar_width = 0.3
bar_pos1 = np.arange(n_subs)
bar_pos2 = bar_pos1 + bar_width
```

```python
year = 2008
#plot graphs for all years
for i in range(len(df_years)):
    plt.figure(figsize=(20,3))
    # Plotting a bar graph

    plt.bar(bar_pos1, df_years[i]['n_comments'].head(n_subs), width=bar_width, label='data')
    plt.bar(bar_pos2, df_years_preproc[i]['n_comments'].head(n_subs), width=bar_width, label='data preprocessed')
    
    #change notation to whole number and rotate subreddit names
    plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
    plt.xticks(bar_pos1 + bar_width / 2, [sub1 + ' | ' + sub2 for sub1, sub2 in zip(df_years[i]['subreddit'].head(n_subs), df_years_preproc[i]['subreddit'].head(n_subs))], rotation='vertical')
    plt.legend()

    # Adding labels and title
    plt.xlabel('subreddits')
    plt.ylabel('n_comments')
    plt.title(f'n_comments per subreddit for {year}')

    # Display the plot
    plt.show()
    year += 1

#plot graphs over all years
plt.figure(figsize=(20,3))

# Plotting a bar graph
plt.bar(bar_pos1, df_total['n_comments'].head(n_subs), width=bar_width, label='data')
plt.bar(bar_pos2, df_total_preproc['n_comments'].head(n_subs), width=bar_width, label='data preprocessed')

#change notation to whole number and rotate subreddit names
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
plt.xticks(bar_pos1 + bar_width / 2, [sub1 + ' | ' + sub2 for sub1, sub2 in zip(df_total['subreddit'].head(n_subs), df_total_preproc['subreddit'].head(n_subs))], rotation='vertical')
plt.legend()

# Adding labels and title
plt.xlabel('subreddits')
plt.ylabel('n_comments')
plt.title(f'n_comments per subreddit total')
```
